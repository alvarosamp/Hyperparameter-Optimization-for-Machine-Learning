```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from skopt.space import Real,Integer,Categorical
```

```python
param = Integer(10, 120, prior = 'uniform', name = 'example')
```

```python
param.rvs()[0]
```

```python
def sample_and_plt(space, title):
    vals_ls = []
    for i in range(500):
        v = space.rvs()
        v = v[0]
        vals_ls.append(v)

    pd.Series(vals_ls).hist(bins = 50)
    plt.title(title)
    plt.ylabel('Number of draws')
    plt.show()

    print('example values: ', vals_ls[0:5])
```

```python
#Sample integers with uniform distribuition
space = Integer(10, 120, prior = 'uniform', name = 'example')
sample_and_plt(space, 'Integer - uniform')
```

```python
#Sample integers with log-uniform distribuition
space = Integer(10, 120, prior = 'log-uniform', name = 'example')
sample_and_plt(space, 'Integer - uniform')
```

```python
#Sample reals with uniform distribuition
space = Real(0.00001, 0.1, prior = 'uniform', name = 'example')
sample_and_plt(space, 'Real - uniform')
```

```python
#Sample reals with log-uniform distibution
space = Real(0.00001, 0.1, prior="log-uniform", name="example")

sample_and_plt(space, 'Real - log-uniform')
```

```python
space = Categorical(['A', 'B', 'C'], name="example")

sample_and_plt(space, 'Categorical')
```
